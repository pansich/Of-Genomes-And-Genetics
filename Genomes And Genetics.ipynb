```python
import pandas as pd
import numpy as np

sample = pd.read_csv('sample_submission.csv')
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
```

```python
full = pd.concat([test, train], axis = 0)
full.columns
```

```python
# Replace -99 with Nan

full.replace(-99, 'NaN', inplace = True)
full
```

```python
full.columns = full.columns.str.lower().str.replace(' ', '_')
full.columns = full.columns.str.replace('heart_rate_(rates/min', 'heart_rate_(rates/min)')
full.columns
```

```python
full = full.drop_duplicates(ignore_index = True)
```
